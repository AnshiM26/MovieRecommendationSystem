```python
import pandas as pd
movies=pd.read_csv('tmdb_5000_movies.csv');
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
credits.head()
```

```python
credits.shape
```

```python
movies=movies.merge(credits, on="title")
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
#columns taken into account 
#genres
#id
#keywords
#title
#cast
#crew
#overview
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
#removing all the null and duplicate values
```

```python
movies.dropna(inplace=True)
```

```python
#Making a new dataframe with three columns that contains movie_id, title and tags. Tags will contain the overview+genres+keywords+cast+crew
#First these 5 columns need to be converted in the format of a list, then it will be appended together and the final list will be converted to a string
#For cast, taking only the first 3 actors, and for crew only the director
#tag column will contain a string paragraph with all the four columns 
```

```python
#function to convert the dictionary to list 
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#To convert cast dictionary to list (taking only the first three actors)
def convertCast(obj):
    count=0
    L=[]
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convertCast)
```

```python
#To convert crew dictionary to list (taking only the name of the director)
def convertCrew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(convertCrew)
```

```python
movies.head()
```

```python
#converting overview string to list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
#Removing the white spaces
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
df=movies[['movie_id','title','tags']]
```

```python
#converting tags list to string (all lowercase)
df['tags']=df['tags'].apply(lambda x:" ".join(x))
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
#Final dataframe
df.head()
```

```python
#Performing stemming on the words in each tag
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split(): #convert the text to a list
        y.append(ps.stem(i))
    return " ".join(y)  #convert list to string
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
#Creating a bag of words model
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',max_features=5000)
```

```python
#Creating the vectors for each movie
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
#Finding out the cosine similarity for the vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
#Algorithm for recommending movies:
#create a function called recommend and take the movie name as a function parameter 
#Find the index of the requested movie
#From the index, fetch the similarity array of that movie
#change the similarity array in the form (index, similarity)
#sort it in reverse (descending order) on the basis of cosine similarity and recommend the first five movies
```

```python
def recommend(movie):
    movie_index=df[df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Titanic')
```
